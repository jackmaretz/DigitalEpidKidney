```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = pd.read_csv("data/output/output_complete.csv",sep=";")
```

```python
X = df.drop("Data_Value", axis = 1) 
y = df.Data_Value
X = X.apply(pd.to_numeric, errors='coerce')
y = y.apply(pd.to_numeric, errors='coerce')
X.fillna(0, inplace=True)
y.fillna(0, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
## The line / model
plt.figure(figsize=(14,10))
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.show()
```

```python
from sklearn.cross_validation import cross_val_score, cross_val_predict
from sklearn import metrics
```

```python
# Make cross validated predictions
predictions = cross_val_predict(model, df, y, cv=6)
plt.scatter(y, predictions)
```

```python
from sklearn.linear_model import Ridge
ridgeReg = Ridge(alpha=0.05, normalize=True)
ridgeReg.fit(X_train,y_train)
pred = ridgeReg.predict(X_test)

#calculating mse

mse = np.mean((pred - y_test)**2)

mse #1348171.96 ## calculating score ridgeReg.score(x_cv,y_cv) 0.5691




```

```python
ridgeReg.score(X_test,y_test)
```

```python
from sklearn.linear_model import Lasso
lassoReg = Lasso(alpha=0.3)

lassoReg.fit(X_train,y_train)

pred2 = lassoReg.predict(X_test)

# calculating mse

mse = np.mean((pred - y_test)**2)

print("mse;",mse)

#1346205.82

lassoReg.score(X_test,y_test)

```

```python
## The line / model
plt.figure(figsize=(14,10))
plt.scatter(y_test, pred)
plt.scatter(y_test, pred2,color="green")


plt.scatter(y_test, predictions,color="red")



plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.legend(["ridge","lasso","lm"])
plt.show()
```

```python
#checking the magnitude of coefficients
from sklearn.linear_model import LinearRegression


lreg = LinearRegression()
lreg.fit(X_train,y_train)
predictors = X_train.columns


coef = pd.Series(lreg.coef_,predictors).sort_values()
plt.figure(figsize=(20,10))


coef.plot(kind='bar', title='Modal Coefficients')
plt.show()
```
